```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Sample data (you can replace it with your own data from sources like Kaggle)
data = {
    'Country': ['USA', 'China', 'India', 'Germany', 'UK', 'Canada', 'Australia', 'Brazil', 'South Africa', 'Japan'],
    'GDP': [65000, 10500, 2200, 55000, 45000, 49000, 60000, 10000, 6000, 41000],
    'LifeExpectancy': [78, 77, 69, 81, 82, 81, 82, 75, 64, 85],
    'Income': [60000, 10000, 3000, 55000, 46000, 50000, 65000, 15000, 6000, 50000],
    'Population': [331, 1441, 1380, 83, 68, 37, 26, 211, 59, 126]
}

# Create DataFrame
df = pd.DataFrame(data)

# Describe data (summary statistics)
print("Summary Statistics:")
print(df.describe())

# Calculate correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
correlation = numeric_df.corr()

## Calculate statistical moments for 'GDP'
mean_value = df['GDP'].mean()  # Assigning the correct calculation for mean
variance_value = df['GDP'].var()  # Variance calculation
skewness_value = stats.skew(df['GDP'])  # Skewness calculation
kurtosis_value = stats.kurtosis(df['GDP'])  # Kurtosis calculation

# Printing the statistical moments
print("\nStatistical Moments for GDP:")
print(f"Mean: {mean_value}, Variance: {variance_value}, Skewness: {skewness_value}, Kurtosis: {kurtosis_value}")

# --- 1. Relational Plot (Scatter plot) ---
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GDP', y='LifeExpectancy', data=df)
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life Expectancy (years)')
plt.show()

# --- 2. Categorical Plot (Bar plot) ---
plt.figure(figsize=(10, 6))
df.groupby('Country')['Income'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Income by Country')
plt.xlabel('Country')
plt.ylabel('Average Income (USD)')
plt.xticks(rotation=45)
plt.show()

# --- 3. Statistical Plot (Correlation Heatmap) ---
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# --- 4. Box Plot (Statistical Plot) ---
plt.figure(figsize=(8, 6))
sns.boxplot(x='Country', y='Income', data=df)
plt.title('Income Distribution by Country')
plt.xticks(rotation=45)
plt.show()

# --- 5. Pair Plot (Statistical Plot) ---
sns.pairplot(numeric_df)
plt.suptitle('Pair Plot for Numerical Data', y=1.02)
plt.show()

# --- PEP-8 Compliance ---
# Ensure that function names, variable names, and overall formatting adhere to PEP-8 standards.
```
